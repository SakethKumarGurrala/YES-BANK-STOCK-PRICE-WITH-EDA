## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

### Time series analysis- Line chart

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
plt.plot(data['Date'],data['Close'],color='Green',label='Close Stock Price')
plt.xlabel('Year')
plt.ylabel('Stock Close Price')
plt.title('Yes Bank Stock Closing Price Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

##### Reason for selecting line cahrt

1. The dataset having the stock prices over a period of time. For Time series analysis **Line charts** are the most effective way to visualize trends and patterns in time series data.

2. It provides a clear view of price movements, helping in identifying long-term growth, volatility, and potential crashes.

##### insights from the chart

1. From 2005 to 2018 their is an significant **growth** in stock price. It reaches to the **max value i.e >350**.

2. From 2018 to 2020 their is a **sharp Decline**. It went up to the its **all time low value which is around 14**.

3. Between 2017 to 2019, The stock price is **more Volatile** before the major decline.


##### Business impact


Helpful for Strategical Decision Making, It helps for investors when to enter and exist the market.

### Closing Prie Distribution - Histogram

```python
plt.figure(figsize=(8,4))
plt.hist(data['Close'], bins=15, edgecolor='black')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.title('Closing Price Distribution')
plt.show()
```

##### Reason for selecting Histogram chart

Histogram is the best choice for visualizing frequency distribution of a few data points of one variable

##### insights from the chart

1. Most closing prices are concentrated below the **100 rupees**. The highest frequency of occurence is in **0-50 & 50-100 Range**, suggesting that the stock spent most of its time in these price ranges.

2. Low frequency in higher price ranges.

#### Stock Price Distribution - Box Plot

```python
plt.figure(figsize=(8,4))
sns.boxplot(data[['Open', 'High', 'Low', 'Close']])
plt.xlabel('Price type')
plt.ylabel('Value')
plt.title('Price Distribution')
plt.grid()
plt.show()
```

##### Reason for selecting Box Plot

Box plot is used because it effectively visualize the distribution, and outliers in stock prices.

##### insights found from the chart

1. **Stock volatility:** The presence of long whiskers and outliers inicates periods of high price flucations.

2. **Price range comparision:** The median values suggest how opening, closing, highest, and lowest prices are distributed, helping identify which price type experiences the most variation.

3. **Outliers Detection:** The circles represnt extreme price points, which could be due to market events, earnings reports.

##### Business impact?


1. For investors: Understanding price variablity helps in assessing risk before making long-term investments.

2. For Traders: The presence of frequent outliers or large price ranges mayb indicates profitable short-term trading opportunities.

### Opening Vs Closing price Realtionship - Scatter plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,5))
plt.scatter(data['Open'], data['Close'], color='Red',alpha=0.5)
plt.xlabel('Opening Price')
plt.ylabel('Closing Price')
plt.title('Opening Vs Closing Price')
plt.show()
```

##### Reason for selecting Scatter plot

A scatter plolt is effectively visualize the relationship between opening and closing prices, showing how closely follow each other. It helps to identify trends, correlations, and potential outliers making it useful for understanding stocks price movements.

##### insights from the chart

1. The opening and Closing prices have a strong positive correlation, indicating that stocks generally close near their opening values.

2. Some deviations suggest market volatility where the stock prices expreience significant movement duing the trading session.

##### Business impact


1. Helps traders anticipates price trends and decide whether to hols or sell stocks.

2. Investers can use thes data to assess stock stability and risk before making investment decisions.

### Correlation Heatmap

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data[['Open', 'High', 'Low', 'Close']].corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Stock Prices')
plt.show()
```

##### Reason for selecting Heat map

A correlation heatmap is used to analyze the relationship between different stock price variables—Open, High, Low, and Close prices.

This heatmap helps in:


*   Identifying strong or weak correlations between price points.
*   Understanding how different price levels move together over time.
*   Providing data-driven insights for stock price behavior.








##### insights from the chart

* All correlations are extremely high (>0.98), meaning stock prices move together in a synchronized manner.
* Open vs. High (0.99) and Close vs. High (0.99) show a near-perfect correlation, meaning the highest price of the day is strongly linked to how the stock opens and closes.
* Low vs. Close (1.00) indicates that the closing price is always directly tied to the day's lowest price.
