## 3. ***Data Wrangling***

### Data Wrangling Code

```python
data.info()
```

```python
# Converting Object into datatime datatype.
data['order_date_time']=pd.to_datetime(data['order_date_time'],format='%d/%m/%Y %H:%M')
data['Issue_reported at']=pd.to_datetime(data['Issue_reported at'],format='%d/%m/%Y %H:%M')
data['issue_responded']=pd.to_datetime(data['issue_responded'],format='%d/%m/%Y %H:%M')
data['Survey_response_Date']=pd.to_datetime(data['Survey_response_Date'],format='%d-%b-%y')
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Customer Remarks'].unique()
```

```python
# Replacing the Null vales in categorical columns
data['Customer Remarks'].fillna('No Remark', inplace=True)
data['Customer_City'].fillna('Unknown City', inplace=True)
data['Product_category'].fillna('Unknown Category',inplace=True)
```

```python
data.isnull().sum()
```

```python
#Droping the null value records from the dataset
data.dropna(subset=['Order_id'], inplace=True)
```

```python
#distplot to show the iteam price
plt.figure(figsize=(8,4))
sns.distplot(data['Item_price'])
plt.show()
```

```python
#Replace Null values with median value
data['Item_price']=data['Item_price'].fillna(data['Item_price'].median())
```

```python
#dropping columns having most missing values
drop_col=['order_date_time','connected_handling_time']
data.drop(drop_col,axis=1,inplace=True)
```

```python
data.isnull().sum()
```

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

### Frequency of Customer Issues on Category- Bar graph

```python
plt.figure(figsize=(12, 6))
category_counts = data['category'].value_counts()
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title('Frequency of Customer Issues On Category')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.show()
```

###Reason for selecting line chart:
A bar chart is the best choice for visualizing the distribution of categorical data, such as Product Category.

###insights from the chart:
1. **Return** is the category having most of the issues by the customers.

2. The second highest issues are from the **Order related**.

3. Very minuate issues from the **App/ Website**, and **Onboarding related**.

###Business impacts:
1. The insights gained from analyzing Customer issues on categorie can have a significant positive impact on business operations.

2. Business Growth & Strategic Decision

### Frequency of Customer issues on Product Categories- Bar graph

### Distrubution of Customer support Rquest channels - Pie Chart

###Reason for selecting pie Plot:
1. Pie chart is used because it effectively visualize the distribution of the data.

###insights found from the chart:

### Relationship between CSAT scores and other variables - Box plot

```python

for col in ['channel_name', 'category', 'Product_category']:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x=col, y='CSAT Score', data=data)
    plt.title(f'CSAT Score vs. {col}')
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

##### Reason for selecting Scatter plot

###Reason for selecting Box Plot:

 A Box plolt is effectively visualize the relationship between variables, here CSAT Vs Channel, CSAT  Vs Category , and CSAT Vs Product Catergory. It helps to identify correlations, and potential outliers making it useful for understanding Customer satisifaction Level.



### Correlation Heatmap

```python
# Correlation Heatmap
corr_column=data.select_dtypes(include=['int64','float64']).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_column, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

###Reason for selecting Heat map:
Heatmaps provide a clear, visual representation of correlations between numerical variables, helping identify trends and dependencies.

By analyzing CSAT scores against factors like response time, businesses can pinpoint the biggest impact areas.

Unusual patterns or extreme values stand out clearly, allowing businesses to identify inefficiencies in their customer support process.

###insights from the chart:

3. If CSAT scores remain low even for fast responses, it suggests that response quality is an issue, not just speed.
